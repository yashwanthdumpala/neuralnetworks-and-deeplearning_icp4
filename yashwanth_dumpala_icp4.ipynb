2. Linear Regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
salary_data = pd.read_csv("/content/Salary_Data (1).csv")
```

```python
salary_data.shape
```

```python
X = salary_data.iloc[:, :-1]
y = salary_data.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
plt.scatter(X_train, y_train, label='Train Data', color='blue')
plt.scatter(X_test, y_test, label='Test Data', color='red')
plt.plot(X_test, y_pred, label='Regression Line', color='green')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend()
plt.show()
```

```python
print("Mean Squared Error:", mse)
```

1. Data Manipulation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "/content/data.csv"
df = pd.read_csv(url)
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
null_values = df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
agg_columns = ['Duration', 'Calories']
aggregated_data = df[agg_columns].agg(['min', 'max', 'count', 'mean'])
```

```python
filtered_calories = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
```

```python
filtered_calories_pulse = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
```

```python
df_modified = df.drop(columns="Maxpulse",axis=1)
```

```python
df.drop(columns=['Maxpulse'], inplace=True)
```

```python
df['Calories'] = df['Calories'].astype(int)
```

```python
plt.scatter(df['Duration'], df['Calories'])
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter Plot of Duration vs Calories')
plt.show()
```

```python
print("\nNull Values:\n", null_values)
print("\nAggregated Data:\n", aggregated_data)
print("\nFiltered Calories:\n", filtered_calories)
print("\nFiltered Calories and Pulse:\n", filtered_calories_pulse)
print("\nModified DataFrame (df_modified):\n", df_modified)
print("\nDataFrame after dropping Maxpulse column:\n", df)
```
